### Sample images from CIFAR100 and CIFAR10 dataset

```python
import numpy as np
import matplotlib.pyplot as plt

from keras.utils import to_categorical

from tensorflow.keras.datasets import cifar100
from tensorflow.keras.datasets import cifar10
```

```python
def read_data(cifar_dataset):

    if cifar_dataset == 100:
        (x_train, y_train), (x_test, y_test) = cifar100.load_data()
        # One-hot encoding for 100 classes.
        y_train = to_categorical(y_train, 100)
        y_test = to_categorical(y_test, 100)
    else :
        (x_train, y_train), (x_test, y_test) = cifar10.load_data()
        y_train = to_categorical(y_train, 10)
        y_test = to_categorical(y_test, 10)

    print('Train: X=%s, y=%s' % (x_train.shape, y_train.shape))
    print('Test: X=%s, y=%s' % (x_test.shape, y_test.shape))
    return (x_train, y_train), (x_test, y_test) 
```

### CIFAR100 dataset

```python
cifar_dataset = 100
(x_train, y_train), (x_test, y_test) = read_data(cifar_dataset)

# plot first few images
for i in range(9):
    #plt.subplot(330 + 1 + i) # encoded as a single integer 3 rows, 3 columns, ith plot
    plt.subplot(3,3,1+i) # subplot 3 rows, 3 columns, ith subplot
    plt.imshow(x_train[i], cmap=plt.get_cmap('gray'))
plt.show()
```

### CIFAR10 dataset

```python
cifar_dataset = 10
(x_train, y_train), (x_test, y_test) = read_data(cifar_dataset)

# plot first few images
for i in range(9):
    #plt.subplot(330 + 1 + i) # encoded as a single integer 3 rows, 3 columns, ith plot
    plt.subplot(3,3,1+i) # subplot 3 rows, 3 columns, ith subplot
    plt.imshow(x_train[i], cmap=plt.get_cmap('gray'))
plt.show()
```
